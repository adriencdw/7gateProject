```
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Load image, grayscale, Gaussian blur, threshold
image = cv2.imread('11_mask.png')
image_original = cv2.imread('11_original.png')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#thresh = cv2.threshold(gray, 220, 255, cv2.THRESH_BINARY_INV)[1]

# Find contours
cnts = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
c = max(cnts, key=cv2.contourArea)

# Obtain outer coordinates
left = tuple(c[c[:, :, 0].argmin()][0])
right = tuple(c[c[:, :, 0].argmax()][0])
top = tuple(c[c[:, :, 1].argmin()][0])
bottom = tuple(c[c[:, :, 1].argmax()][0])


# Draw dots onto image

min_point=(left[0] ,top[1]  )
max_point=(right[0] ,bottom[1])

cv2.rectangle(gray,min_point,max_point,(255,255,255),3)


print('left: {}'.format(left))
print('right: {}'.format(right))
print('top: {}'.format(top))
print('bottom: {}'.format(bottom))

cv2.rectangle(image_original,min_point,max_point,(255,0,0),3)


plt.imshow(image)

plt.show()

plt.imshow(gray)

plt.show()



plt.imshow(image_original)

plt.show()
```
